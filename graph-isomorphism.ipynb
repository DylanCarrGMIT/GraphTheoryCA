#What is Graph Isomorphism?
Graph isomorphism is a "bijection"(or one-to-one correspondence) between the vertex sets of two graphs. Basically, if two graphs have the same number of vertices and edges, and still retain their edge connectivity, then they are isomorphic. [1]



#The Graph Isomorphism Problem
This is the computational problem regarding algorithmically determining whether two graphs are isomorphic or not. In complexity terms, computational problems are broadly categorised into 6 different groups, the relevant two regarding graph isomorphism being "P", meaning completable in polynomial time on a deterministic Turing machine, and "NP", meaning it can be solved in polynomial on a non-deterministic Turing machine. [2]

The graph isomorphism problem is unique, as it has yet to be found to be solvable in polynomial time, yet also is not considered "NP-complete". Because of this, it has eluded a very clear classification for decades.
