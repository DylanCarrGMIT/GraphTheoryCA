#What is Graph Isomorphism?
Graph isomorphism is a "bijection"(or one-to-one correspondence) between the vertex sets of two graphs. Basically, if two graphs have the same number of vertices and edges, and still retain their edge connectivity, then they are isomorphic. [1]



#The Graph Isomorphism Problem
This is the computational problem regarding algorithmically determining whether two graphs are isomorphic or not. In complexity terms, computational problems are broadly categorised into 6 different groups, the relevant two regarding graph isomorphism being "P", meaning completable in polynomial time on a deterministic Turing machine, and "NP", meaning it can be solved in polynomial on a non-deterministic Turing machine. [2]

#Computational Complexity
The graph isomorphism problem is unique, as it has yet to be found to be solvable in polynomial time, yet also is not considered "NP-complete". Because of this, it has eluded a very clear classification for decades. So much so that a new classification was created for it, "GI". [3]

#Representing Graphs in Data Structures
Graphs can be represented in 3 different types of data structure: [4]
1. Adjacency Matrix
2. Adjacency List
3. Adjacency Set

#Adjacency Matrix
An adjacency matrix is like a table, consisting of rows and columns. By creating a table in this way, you can show a connection by marking applicable fields with a "1", and fields with no connections with a "0".



Example of Adjacency Matrix[4]
#Adjacency List
Adjacency lists work by representing each vertex as a node, which contains either data, or a reference to a linked list, the linked list listing all adjacent nodes. Adjacency lists are faster to search through than an adjacency matrix, though removing a node would require searching all lists to find the node for removal.


Example of Adjacency List [4]

#Adjacency Set
Adjacency sets are nearly identical to adjacency lists, just that it uses a set of adjacent vertices instead of a linked list.
Generally speaking, when it comes to selecting a data structure for graphs, adjacency matrixes are the better choice for well-connected graphs with many nodes, while adjacency sets/lists work better for more sparsely populated graphs.
